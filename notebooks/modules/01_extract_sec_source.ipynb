# Module 01 — Extract SEC Source Data

This module only handles source discovery and ingestion.
No API key is required here.

```python
from pathlib import Path
import pandas as pd

candidate_paths = [
    Path('/Users/game./Documents/Sandbox/sec_api_data.csv'),
    Path('/Users/game./Documents/Sandbox/bond_features_by_company.csv'),
    Path('/Users/game./Documents/Sandbox/bond_features_cleaned.csv'),
]

source_path = next((p for p in candidate_paths if p.exists()), None)

if source_path is None:
    print('No source CSV found; creating a demo dataset for reproducibility.')
    df_raw = pd.DataFrame({
        'company_id': ['C0000008240', 'C0000008240', 'C0000033672'],
        'thaibma_symbol': ['DEMO26A', 'DEMO27A', 'DEMO28A'],
        'maturity__issue_date': ['2026-05-05', '2026-05-20', '2027-01-10'],
        'coupon__rate': ['3.10', '3.40', '4.25'],
        'credit_rating': ['A', 'A-', 'BBB+'],
    })
    source_name = 'demo-generated'
else:
    print(f'Loading source from: {source_path}')
    df_raw = pd.read_csv(source_path, low_memory=False)
    source_name = str(source_path)

print(f'Rows: {len(df_raw):,} | Columns: {df_raw.shape[1]:,}')
display(df_raw.head(8))
```

```python
missing_ratio = (df_raw.isna().mean() * 100).sort_values(ascending=False).head(10)
print('Top missing columns (%):')
display(missing_ratio.to_frame('missing_pct'))
```

```python
out_dir = Path('/Users/game./Documents/Sandbox/Bond_starter_project_public/data/processed')
out_dir.mkdir(parents=True, exist_ok=True)
raw_out = out_dir / 'sec_bond_raw_snapshot.csv'
df_raw.to_csv(raw_out, index=False)
print(f'Saved raw snapshot: {raw_out}')
print(f'Source used: {source_name}')
```
