# Module 03 — Bond Screening (Issue Month = 05/2026)

This module performs a focused screening request using cleaned data.

```python
from pathlib import Path
import pandas as pd
```

```python
cleaned_path = Path('/Users/game./Documents/Sandbox/Bond_starter_project_public/data/processed/sec_bond_cleaned.csv')
if not cleaned_path.exists():
    raise FileNotFoundError('Run module 02 first to create sec_bond_cleaned.csv')

df = pd.read_csv(cleaned_path, low_memory=False)

date_col = 'maturity__issue_date' if 'maturity__issue_date' in df.columns else None
if date_col is None:
    raise KeyError('No maturity__issue_date column found in cleaned data.')

df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
screened_df = df[(df[date_col].dt.year == 2026) & (df[date_col].dt.month == 5)].copy()
screened_df['issue_month_display'] = screened_df[date_col].dt.strftime('%m/%Y')

print(f'Screened rows (05/2026): {len(screened_df):,}')
display(screened_df.head(20))
```

```python
out_dir = Path('/Users/game./Documents/Sandbox/Bond_starter_project_public/data/processed')
out_dir.mkdir(parents=True, exist_ok=True)
screened_path = out_dir / 'sec_bond_screened_2026_05.csv'
screened_df.to_csv(screened_path, index=False)
print(f'Saved screened dataset: {screened_path}')
```
